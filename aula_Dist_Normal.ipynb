```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
med = 7
std = 2
med2 = 7
std2 = 1.5
x = np.linspace(1,13)

plt.figure(figsize=(10,8))
plt.plot(x, norm.pdf(x,med,std),
        label = '$\mathcal{N}$  ' + f'$( \mu = {med:.2f} , \sigma = {std:.2f} )$')
plt.plot(x, norm.pdf(x,med2,std2), 'r',
        label = '$\mathcal{N}$  ' + f'$( \mu = {med2:.2f} , \sigma = {std2:.2f} )$')
plt.title('Distribuição Normal')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```

```python
med = 0
std = 1
med2 = 0
std2 = 2.0
med3 = 2
std3 = 1
x = np.linspace(-5,5)

plt.figure(figsize=(10,8))
plt.plot(x, norm.pdf(x,med,std),
        label = '$\mathcal{N}$  ' + f'$( \mu = {med:.2f} , \sigma = {std:.2f} )$',linewidth=5.0)
plt.plot(x, norm.pdf(x,med2,std2), 'r',
        label = '$\mathcal{N}$  ' + f'$( \mu = {med2:.2f} , \sigma = {std2:.2f} )$',linewidth=5.0)
plt.plot(x, norm.pdf(x,med3,std3), 'g',
        label = '$\mathcal{N}$  ' + f'$( \mu = {med3:.2f} , \sigma = {std3:.2f} )$',linewidth=5.0)
plt.ylim(0,0.45)
plt.title('Distribuição Normal Padrão')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```
